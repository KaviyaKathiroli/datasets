# machine learning model

```python
# performing categorical variable to numeric variable with label encoder to make uniform data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Sex']= le.fit_transform(data['Sex'])
data.head()
```

# Outliers

```python
from scipy import stats
abalone_data_outlier= data.copy()
```

```python
abalone_data_outlier
```

```python
z = np.abs(stats.zscore(abalone_data_outlier))
print(z)
```

```python
a=np.where(z > 3)
```

```python
a
```

```python
abalone_data_outlier1= abalone_data_outlier[(z < 3).all(axis=1)]
```

```python
abalone_data_outlier1
```

```python
abalone_data_outlier1.shape
```

```python
data.shape
```

```python
X = abalone_data_outlier1.drop('Rings', axis = 1) # Input/independent variable
y = abalone_data_outlier1. Rings
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
pred = rfr.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error
print('Root Mean Squared Error is: ', np.sqrt(mean_squared_error(y_val, pred)))
```
