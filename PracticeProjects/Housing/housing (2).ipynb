Dropping non related columns

```python
data.drop(['Id'],axis=1,inplace=True)
data.drop(['MSSubClass'],axis=1,inplace=True)
data.drop(['MSZoning'],axis=1,inplace=True)
data.drop(['LotFrontage'],axis=1,inplace=True)
data.drop(['Alley'],axis=1,inplace=True)
data.drop(['GarageCars'],axis=1,inplace=True)
data.drop(['MiscFeature'],axis=1,inplace=True)
data.drop(['MiscVal'],axis=1,inplace=True)
```

Seggregating features

```python
numerical_feature=[feature for feature in data.columns if data[feature].dtypes!='O']
descrete_feature=[feature for feature in numerical_feature if len(data[feature].unique())<25]
cont_feature=[feature for feature in numerical_feature if feature not in descrete_feature]
category_feature=[feature for feature in data.columns if feature not in numerical_feature]
```
