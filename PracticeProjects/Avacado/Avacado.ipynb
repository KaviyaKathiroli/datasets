```python
def get_uniques(df,columns):
    return{column: list(df[column].unique()) for column in columns}
```

```python
catergories=['region','Date','type']
get_uniques(avacadoset,catergories)
```

```python
ordinal_feature=['Date']
nominal_feature=['region']
target_column='type'
```

```python
Date_order=sorted(avacadoset['Date'].unique())
```

```python
pd.get_dummies(avacadoset['region'])
```

```python
Date_order.index('2018-02-04')
```

```python
def ordinal_encode(df,column,ordering):
    df=df.copy()
    df[column]=df[column].apply(lambda x:ordering.index(x))
    return df
```

```python
def onehot_encode(df,column):
    df=df.copy()
    dummies=pd.get_dummies(df[column])
    df=pd.concat([df,dummies],axis=1)
    df.drop(column,axis=1,inplace=True)
    return df
```

```python
avacadoset=ordinal_encode(avacadoset,'Date',Date_order)
```

```python
avacadoset=onehot_encode(avacadoset,'region')
```

```python
avacadoset
```

```python
Label_encoder=LabelEncoder()
avacadoset[target_column]=Label_encoder.fit_transform(avacadoset[target_column])
```

```python
avacadoset.info()
```

```python
y=avacadoset[target_column]
X=avacadoset.drop(target_column,axis=1)
```

```python
X
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
y.shape
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
```

```python
y.sum()/len(y)
```

```python
inputs=tf.keras.Input(65,)
x=tf.keras.layers.Dense(64,activation='relu')(inputs)
x=tf.keras.layers.Dense(64,activation='relu')(x)
outputs=tf.keras.layers.Dense(1,activation='sigmoid')(x)

model=tf.keras.Model(inputs=inputs,outputs=outputs)

model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
batch_size=64
epochs=100
history=model.fit(X_train,y_train,validation_split=0.2,batch_size=batch_size,epochs=epochs
                  ,callbacks=[tf.keras.callbacks.ReduceLROnPlateau()],verbose=2)
```

```python
plt.figure(figsize=(14,10))
epochs_range=range(1,epochs+1)
train_loss=history.history['loss']
test_loss=history.history['val_loss']

plt.plot(epochs_range,train_loss,label="Training_Loss")
plt.plot(epochs_range,test_loss,label="Validation_loss")

plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.show()
```
